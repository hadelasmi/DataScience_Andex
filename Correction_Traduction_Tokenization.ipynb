```python
from textblob import TextBlob,Word
```

```python
with open("C:/Users/lenovo/Desktop/data.txt",'r',encoding='utf-8') as r:  
    with open("C:/Users/lenovo/Desktop/translated_1.txt",'w') as n:
        for line in r:
            #print(line)
            fr_blob = TextBlob(line)
            R = fr_blob.translate(from_lang="fr", to='en')
            n.write(str(R))
```

```python
#Module de tokenization du texte traduit:
```

```python
import re
from nltk.corpus import stopwords 
from textblob import TextBlob, Word
stop_words = set(stopwords.words("english")) 
filtered_sentence =[]
with open("C:/Users/lenovo/Desktop/translated_1.txt",'r', encoding='utf-8') as f2:
    data = f2.read()
s = re.sub(r'[^\w\s]','',data)
v1 = s.split(sep=' ')
sent = TextBlob(s)
Last = " ". join([w.lemmatize() for w in sent.words])
v1 = Last.split(sep=' ')
for w in v1:
    if w not in stop_words and len(w) >1 :
        filtered_sentence.append(w)
mylist = list(dict.fromkeys(filtered_sentence))
print(mylist)
```
